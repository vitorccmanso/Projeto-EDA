## 1 - Carregar os arquivos e cadastro, investimento1 e investimento2 e mostrar a quantidade de linhas de cada um, além das informações desses dataframes

```python
import pandas as pd
import datetime as dt
import os

dir_atual = os.getcwd()
caminho_cadastro = os.path.join(dir_atual, 'arquivos', 'cadastro.csv')
caminho_investimento1 = os.path.join(dir_atual, 'arquivos', 'investimento_parte1.csv')
caminho_investimento2 = os.path.join(dir_atual, 'arquivos', 'investimento_parte2.csv')

df_c = pd.read_csv(caminho_cadastro)
df_inv1 = pd.read_csv(caminho_investimento1)
df_inv2 = pd.read_csv(caminho_investimento2)

print(f"Quantidade de linhas do arquivo 'cadastro': {len(df_c)}")
print(f"Quantidade de linhas do arquivo 'investimento_1': {len(df_inv1)}")
print(f"Quantidade de linhas do arquivo 'investimento_2': {len(df_inv2)}")
print('-'*64)

print("Informações sobre o arquivo 'cadastro':")
print(df_c.info())
print('-'*64)

print("Informações sobre o arquivo 'investimento_1':")
print(df_inv1.info())
print('-'*64)

print("Informações sobre o arquivo 'investimento_2':")
print(df_inv2.info())
print('-'*64)
```

## 2 - Juntar os dataframes de investimento1 e investimento2 em um só e ver a quantidade de linhas

```python
df_inv = pd.concat([df_inv1, df_inv2])
print(f"Quantidade de linhas: {len(df_inv)}")
df_inv
```

## 3 - Crie um novo dataframe eliminando as linhas duplicadas

```python
duplicatas = df_inv.duplicated()
df_inv = df_inv[~duplicatas] #O sinal '~' é utilizado como um 'NOT', selecionando as linhas não duplicadas
print(f"Quantidade de linhas: {len(df_inv)}")
df_inv
```

## 4 - Junte o dataframe de investimento com o de cadastro
Dica: use a chave id

```python
df = pd.merge(df_c, df_inv, on='id')
df
```

## 5 - Ordene o dataframe pelo id, de forma crescente e redefina o índice

```python
df = df.sort_index()
df = df.set_index('id')
df
```

## 6 - Faça uma análise dos dados faltantes e trate-os

Mostre todas as colunas com dados faltantes, sua quantidade/porcentagem e como tratar (preencher) cada uma delas explicando sua estratégia

```python
colunas_dados_faltantes = df.isna() #Retorna o dataframe cheio de booleanos. Ao utilizar o método 'any' em conjunto, retorna todas as colunas e um booleano ao lado
quantidade_faltantes = colunas_dados_faltantes.sum()
porcentagem_faltantes = (quantidade_faltantes / len(df)) * 100
infos_faltantes = pd.concat([quantidade_faltantes, porcentagem_faltantes.round(2)], axis=1, keys=["Quantidade", "Porcentagem"]) #Concatena 'quantidade_faltantes' e 'porcentagem_faltantes' pelo eixo de colunas, adicionando nomes para melhor visualização
print(infos_faltantes[infos_faltantes["Quantidade"] > 0]) #Exibe apenas as colunas do dataframe que possuem 'NaN', a quantidade e porcentagem
print("Nota NaN na coluna Nota1 e Nota2:")
print(len(df[df['Nota1'].isna() & df['Nota2'].isna()]))
print("Nota NaN na coluna Nota1 ou Nota2:")
print(len(df[df['Nota1'].isna() | df['Nota2'].isna()]))

```

### As colunas que possuem NaN são as colunas dos ativos 1 ao 6 e das notas. No caso dos ativos, como indica apenas a quantidade investida por investidor, os valores NaN podem ser substituidos por 0, indicando que o investidor não possui o ativo em questão.
### Já nas colunas das notas, os valores NaN podem ser substituidos por números negativos, para que quando as colunas da média e perfil NPS de cada investidor sejam criadas, um filtro possa ser estabelecido para aqueles que possuem média negativa, indicando que aquele investidor não deu notas suficientes para entrar na métrica do NPS. Nesse caso, uma categoria extra será criada para indicar que aquele investidor não deu a quantidade de notas suficientes para entrar no calculo do NPS.

```python
df.loc[:, 'Ativo01':'Ativo06'] = df.loc[:, 'Ativo01':'Ativo06'].fillna(0)
df.loc[:, 'Nota1':'Nota2'] = df.loc[:, 'Nota1':'Nota2'].fillna(-101)
df.isna().any()
```

## 7 - Manipule os dados a seguir:
- Criar coluna "Total" somando o valor de todos os ativos por investidor
- Criar coluna "Media" com a média das notas 1 e 2
- Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)
- Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)
- Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar o map)
- Transformar os dados da coluna Media em inteiro
- Criar coluna "NPS" onde o investidor é promotor se a média das notas é maior igual a 80, neutro se a média estiver entre 50 e 80, e detrator se for abaixo de 50


```python
df["Total"] = df.loc[:, 'Ativo01':'Ativo07'].sum(axis=1)
df["Média"] = df[['Nota1', 'Nota2']].mean(axis=1)
df["Média"] = df["Média"].round().astype("int")
df["Dominio"] = df["Email"].str.split('@').str[1]
```

```python
df = df.rename(columns={'Salario': 'Renda'})
df["Renda"] = df['Renda'].str.replace("$", "").astype("float")
```

```python
df["Sexo"] = df["Sexo"].map(lambda x: 'Masculino' if x == 'Masc' else 'Feminino')
```

```python
def nps(media):
    if media >= 0:
        if media >= 80:
            return 'Promotor'
        elif media < 50:
            return 'Detrator'
        else:
            return 'Neutro'
    return 'Falta Nota'

df['NPS'] = df["Média"].apply(lambda media: nps(media))
df
```

## 8 - Calcule os seguintes dados
- Calcule o total investido em cada coluna Ativo (tente usar o apply)
- Qual o menor valor investido em cada ativo?
- Mostre a média e desvio padrão de cada ativo
- Qual a idade do investidor mais velho?
- Qual a mediana da coluna Média?

```python
import numpy as np
pd.options.display.float_format = '{:.2f}'.format  # formatação da notação científica

ativos = df.loc[ : , 'Ativo01':'Ativo07']

# Total investido em cada coluna Ativo
total_por_ativo = ativos.apply(np.sum)
print("Total investido em cada Ativo: ")
display(total_por_ativo)
print("-"*32)


# Menor valor investido em cada Ativo
menor_por_ativo = ativos.loc[(ativos!=0).all(axis=1)].apply(np.min)  # remoção dos valores zerados
print("Menor valor investido em cada Ativo: ")
display(menor_por_ativo)
print("-"*32)


# Média e desvio-padrão de cada Ativo
media_por_ativo = ativos.apply(np.mean).round(2)
print("Média de investimento em cada Ativo: ")
display(media_por_ativo)
print("-"*32)

desvio_por_ativo = ativos.apply(np.std).round(2)
print("Desvio-padrão em cada Ativo: ")
display(desvio_por_ativo)
print("-"*32)


# Idade do investidor mais velho
investidor_mais_velho = df.loc[ : , 'Idade'].max()
print(f"Idade do investidor mais velho: \n{investidor_mais_velho}")
print("-"*32)


# Mediana da coluna Média
mediana_media_nps = df.loc[ : , 'Média'].median()
print(f"Mediana da coluna Média de NPS: \n{mediana_media_nps}")
print("-"*32)
```
